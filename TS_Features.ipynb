```python
from ts_features import tsf_vectorizer
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/ivanfilhoreis/ts_features/main/dataset/soja_cbot.csv")
df['Date'] = pd.to_datetime(df['Date'])
df['Volume'].fillna(method = 'ffill', inplace = True)
df.sort_values(by='Date', inplace = True)
df = df[(df['Date'] > '2015-01-01') & (df['Date'] < '2023-01-01')]
df.reset_index(drop=True, inplace=True)

df.head()
```

## Features in percentage difference

Extract features in percentage difference of each time series point in relation to its components.


```python
ts = tsf_vectorizer(feature='perc')

df_tsf = ts.fit_transform(df)
```

```python
df_tsf.sample(20)
```

## Features on labels

It assigns label to each feature:

- value positive: 1
- value negative: -1
- insignificant difference: 0

```python
ts_lb = tsf_vectorizer(feature='label')

tsf_lb = ts_lb.fit_transform(df)
```

```python
tsf_lb.sample(20)
```

## Extract features value

It extracts the value difference for each component of the time series.

```python
ts_vl = tsf_vectorizer(feature='value')

tsf_vl = ts_vl.fit_transform(df)
```

```python
tsf_vl.sample(10)
```

# Atribute label from time series features

```python
ts_pr = tsf_vectorizer()

tsf_pr = ts_pr.extract_label(df)
```

```python
tsf_pr.sample(20)
```

```python
tsf_pr2 = ts_pr.extract_label(df, bin=(0, -1))

tsf_pr2.sample(20)
```
